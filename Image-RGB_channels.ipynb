```python
import imageio
import matplotlib.pyplot as plt
%matplotlib inline

# Read image
im = imageio.imread("./Figures/chapel.png")
print(f"Shape of the image: {im.shape}")

plt.figure()
plt.axis('off')
plt.imshow(im)
plt.show()
```

```python
# Red channel
plt.figure()
plt.axis('off')
plt.imshow(im[:,:,0], cmap='gray')
plt.show()
```

```python
# Green channel
# It's easy to see the difference on the sky
plt.figure()
plt.axis('off')
plt.imshow(im[:,:,1], cmap='gray')
plt.show()
```

```python
# Blue channel
plt.figure()
plt.axis('off')
plt.imshow(im[:,:,2], cmap='gray')
plt.show()
```
