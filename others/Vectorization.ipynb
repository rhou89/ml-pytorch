```python
# Exmaple showing SIMD advantage even on a CPU
# TODO: A full benchmark on this would be interesting.

import matplotlib.pyplot as plt
import numpy as np
import time
```

```python
def doit(length):
    a = np.random.rand(length)
    b = np.random.rand(length)

    tic = time.time()
    c = np.dot(a, b)
    toc = time.time()
    time1 = (toc-tic)*1000

    c = 0
    tic = time.time()
    for i in range(length):
        c += a[i]*b[i]
    toc = time.time()
    time2 = (toc-tic)*1000

    return time1, time2
```

```python
tot = 8
benchmark = np.array([doit(10**x) for x in range(tot)])

plt.title('Computation time comparsion in log scale')
plt.xlabel('log10(Vector size)')
plt.xlabel('log10(Computation time (ms))')
plt.plot(range(tot), np.log10(benchmark[:,0]), label = 'Vectorization')
plt.plot(range(tot), np.log10(benchmark[:,1]), label = 'For loop')
plt.legend()
plt.show()
```
